No hay valores faltantes, pero veo que hay muchos columnas con el caracter "?" que no corresponde a ningun dato especifico, se intentara reemplazar con valor nulo ya que reemplazarlo por 0 seria un grave error debido a que en algunas columnas es un dato de interes. 

```python



question_marks_count = (df == "?").sum()


columns_with_question_marks = question_marks_count[question_marks_count > 0]


print("Columnas con '?' y la cantidad de veces que aparecen:")
print(columns_with_question_marks)
```

```python
df.replace('?', np.nan, inplace=True)
```

Ahora chequeamos los valores faltantes, luego veremos como tratar a estos valores

```python

nan_counts = df.isnull().sum()


print("Cantidad de valores NaN en cada columna:")
print(nan_counts)
```

```python
df[df.duplicated()]
```

Ahora eliminaremos las columnas que a mi criterio no nos sirven para nuestro proyecto, como cn_ln(No pude averiguar que es y el nombre no es muy espeficio por lo que se me es imposible saber que es), Summary(que aunque dice las causas del bueno no he logrado poder extraer datos relevantes cuantificables), flaigth_no(Tiene demasiados valores faltantes como para tenerla en cuenta), registro(Parece ser la matricula del aeronave)

```python
df.drop(['flight_no', 'registration', 'cn_ln', "HORA declarada"], axis=1, inplace=True)
```

Tambien eliminaremos la columna route que es la misma que ruta y tiene mas datos faltantes 

```python
df.drop(["route"], axis=1, inplace=True)
```

Ahora voy a limpiar eliminar los valores NaN de ALL_aboard, PASAJEROS A BORDO, crew_aboard, cantidad de fallecidos, passenger_fatalities, crew_fatalities. De esta manera puedo trabajar de manera mas limpia y simple las variables numericas.
ademas al ser pocos datos los que faltan creo es mas conveniente.
Tambien borraremos HORA_declarada ya que no parece aportar datos a nuestra investigacion

```python



columnas = ['all_aboard', 'PASAJEROS A BORDO', 'crew_aboard', 'cantidad de fallecidos',
            'passenger_fatalities', 'crew_fatalities', 'ground']
df.dropna(subset=columnas, how='any', inplace=True)


df[columnas] = df[columnas].astype(int)

columnas1 = ["Ruta", "OperadOR", "ac_type", "summary"]
df.dropna(subset=columnas1, how='any', inplace=True)
```

Con esto note que hay algunas columnas donde all_abord  y	crew_aboard estan en 0 lo cual es un error a menos que el avion se vuele solo y no lleve a nadie asique eliminare todas las filas que cumplan con la condicion de que halla si o si crew a bordo al menos. si hay en all_abord y	crew_aboard ambas en 0, seran eliminadas y si	crew_aboard es igual a 0 tambien seran eliminadas

```python
df = df.drop(df[(df['all_aboard'] == 0)
                 & (df['crew_aboard'] == 0)].index)
df= df.drop(df[df['crew_aboard'] == 0].index)
df = df[df['all_aboard'] != 0]
```

Ahora crearemos dos columnas nuevas en base a "fecha" que seran "Año_colicion_, "mes_colision y luego borraremos la columna fecha"

```python
df['fecha'] = pd.to_datetime(df['fecha'])

df['Año_colision'] = df['fecha'].dt.year
df['mes_colision'] = df['fecha'].dt.month


df.drop('fecha', axis=1, inplace=True)
df.drop('Id', axis=1, inplace=True)
```

```python
df.head()
```

```python

num_filas = df.shape[0]

print("Número de filas:", num_filas)
```

Corroboramos el tipo de dato de cada columna y corregimos 

```python

tipos_columnas = df.dtypes


print(tipos_columnas)
```

Convertimos las del tipo objets en str 

```python
columnas_str = ['Ruta', 'OperadOR', 'ac_type']
df[columnas_str] = df[columnas_str].astype(str)
```

```python



df["OperadOR Militar"] = df["OperadOR"].apply(
    lambda x: x if "military" in x.lower() else None
)
df["OperadOR no Militar"] = df["OperadOR"].apply(
    lambda x: x if "military" not in x.lower() else None
)


df['OperadOR Militar'] = df['OperadOR Militar'].fillna(0)
df['OperadOR no Militar'] = df['OperadOR no Militar'].fillna(0)


df['OperadOR Militar'] = df['OperadOR Militar'].apply(
    lambda x: 1 if str(x) != '0' and str(x) != '' else x)
df['OperadOR no Militar'] = df['OperadOR no Militar'].apply(
    lambda x: 1 if str(x) != '0' and str(x) != '' else x)
```

```python

df['crashed'] = 0
df['failure'] = 0
df['weather'] = 0
df['other'] = 0


df.loc[df['summary'].str.contains(
    r'\bcrashed\b', case=False, regex=True), 'crashed'] = 1
df.loc[df['summary'].str.contains(
    r'\bfailure\b', case=False, regex=True), 'failure'] = 1
df.loc[df['summary'].str.contains(
    r'\bweather\b', case=False, regex=True), 'weather'] = 1
df.loc[(df['crashed'] == 0) & (df['failure'] == 0)
       & (df['weather'] == 0), 'other'] = 1

df.head(20)
```

Guardamos el archivo 

```python



df = df.reset_index(drop=True)


df['nuevo_id'] = range(len(df), 0, -1)


print(df.head())
df.to_csv('AccidentesAviones_norm1.csv', index=False)
```
